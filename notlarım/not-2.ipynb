# 2.Ders Notu

Python'da değişkenlere atama her değişken memory'de ayrı bir alan tutmaktansa sadece ortak alanı işaret eder mesela bir örnek:

````python
temp = 35
print("temp : ", temp)
sicaklik = 35
print("sicaklik:", sicaklik)

print("temp id = ", id(temp))
print("sicaklik id =",id(sicaklik))
````

Örnekte de görüldüğü gibi farklı değişkenler olmalarına rağmen aynı id'yi veriyor, bu da aynı değeri işaret etmeleri ile sağlanıyor ve böylece daha iyi bir memory yönetimini destekliyor.

Atama operatörü olan `=` işareti verinin o değişkeni işaret eden sığ bir kopyasını (shallow copysini) oluşturur.

Dizgi (string) ler bir sekans türüdür yani sıranın önemli olduğu değerler koleksiyonudur. Listeler (lists) ve demetler (tuples) de sekans türüdür.

### Veri Tipleri

- *str :* Dizgi, düzyazı ver tipidir ve bir sekans veri türüdür.
- *int :* Tamsayı veri tipidir.
- *float :* Kesirli, kayan sayı veri tipidir.
- *bool :* True veya False değerli ikili(binary) mantık veri tipidir.
- *none:* None python'da boş değer yani null değerine karşılık gelen veri tipidir.

*Type Conversion (Tip Çevirme)* işlemi, çevirmek istediğimiz veriyi çevirmek istediğimiz veri tipi fonksiyonu içerisine alırız bu sayede veri tipi dönmüş olur. Örneğin :

````python
example = '2004'
print(type(example))
example_but_int = int(example)
print(type(example_but_int))
````

### Fonksiyonlar

Fonksiyonlar tanımladığımız belirli bazı görevleri yerine getirmemize yarayan işlevsel kod bloklarıdır. Tekrar eden görevler için kolaylık sağlarlar. Parametrelere sahiptirler ve bu parametreler ile değişken durumlarda istenilen görevleri yerine getirirler.

Pythonda önceden tanımlı fonksiyonlar da vardır mesela;
-  `print()` fonksiyonu içerisinde yazdığımız veriyi ekrana yazdırılmasına yarar.
- `id()` fonksiyonu içerisine parametresine yazdığımız değişkenin bellekteki numarasını verir.

Kendi fonksiyonlarımızı yazmamızı sağlayan fonksiyonda `def` fonksiyonudur;

````python
def deneme(username):
    """Kullanıcı adınızı döndürür."""
    print(f"Kullanıcı adınız: {username}'dır/dir/tır/tir.")

deneme("efe")
````

### return Deyimi

Bir fonksiyon kullanıldıktan sonra fonksiyonun ürettiği sonucu tutmak ve bunu kaydetmek istediğimizde yani sadece çalışıp bir değer tutmamasındansa ürettiği sonucu tutmasını istediğimizde fonksiyonun içierisinde bu deyimi kullanırız böylece üretilen sonuçları return edebiliriz yani kullanıcıya döndürebiliriz. Örneğin;

````python
def fonk_not_return(first_name):
    """Kullanıcın ilk ismi."""
    print(f"Kullanıcının ilk ismi: {first_name}")

fonk_not_return("Hasan")

type(fonk_not_return("Hasan"))
````

`fonk_not_return` fonksiyonu içerisindeki parametre ile beraber type sorgusu yaptığımızda herhangi bir tipe uymadığını yani fonksiyonun bu sonucu tutmadığını sadece verdiğimiz işi yaptığını görüyoruz.

Şimdi return deyimini kullanarak fonksiyonun işini yaparken tam olarak neyi döndüreceğini açıkça belirterek bunu tutmasını da sağlarız. Örneğin;

````python
def fonk_return(surname):
    """Kullanıcın soyadı."""
    print(f"Kullanıcının soyadı: {surname}")
    return surname

print(type(fonk_return("kocasu")))

s = fonk_return("kocasu")
print(s)
type(s)
````

return değeri ile dönmesini istediğimiz şeyi fonksiyonda belirtmiş olduk ve fonksiyonumuz döndürmesini istediğimiz şeyi (burada parametreye girilen soyadını döndürmesini ve saklamasını(tutmasını) istedik) belirttik. Ve type ile baktığımızda fonksiyonun çağırdğımız şekliyle str tuttuğunu da açıkça gördük. (FOnksiyonların tanımlarken en üstteki `"""` ile tanımlanan kısım fonksiyonun açıklamasıdır ve ona docstring (document string) denir. `.__doc__` ile görülebilir ve fonksiyonun yaptığı işlem öğrenilebilir. Daha karmaşık fonksiyonlarda bu hayati öneme sahiptir.)(`__ __` ile gösterilen şeyler pythonda dunder methods'dur daha sonra ayrıntılı olarak açıklaaycağım.)

### Anonim Fonksiyonlar (Lambda fonksiyonu)

Bu fonksiyonlar, `def` anahtar sözcüğü kullanılarak standart biçimde deklere edilmedikleri için anonim fonksiyonlar (anonymous functions) olarak adlandırılır. 

Küçük anonim fonksiyonlar oluşturmak için `lambda` anahtar sözcüğünü kullanabilirsiniz.

Bu nedenle anonim fonksiyonlara lambda fonksiyonları da denir. Python'daki bir lambda fonksiyonu aşağıdaki sözdizimine (syntax) sahiptir.

```
lambda arguments: expression
```

Lambda fonksiyonları herhangi bir sayıda argümana sahip olabilir, ancak yalnızca bir ifadeye (expression) sahip olabilir. İfade değerlendirilir ve döndürülür. Lambda fonksiyonları, fonksiyon nesnelerinin gerekli olduğu her yerde kullanılabilir.

Örneğin, bir sayının karesini alan bir fonksiyonu önceden yazmıştır:

````python
def square_(num):
    return num**2

square_(num = 2)
````

Bu fonksiyonu bir lambda fonksiyonu olarak kolaylıkla yazabiliriz:

````python
square_ = lambda x: x ** 2

print(square_(2))
````

Yukarıdaki programda `square_ = lambda x: x * 2`, lambda fonksiyonudur. Burada `x` bir argümandır (parametre,değişken) ve `x * 2`, değerlendirilen ve döndürülen ifadedir.

Yukarıda da dediğimiz gibi Lambda fonksiyonları herhangi bir sayıda argümana sahip olabilir, ancak yalnızca bir ifadeye (expression) sahip olabilir.

````python
def iki_sayi_toplami(a, b):
    return a + b

iki_sayi_toplami(a = 8, b = 4)
````

````python
iki_sayi_toplami = lambda a, b: a + b

iki_sayi_toplami(8, 4)
````

Lambda fonksiyonlarının diğer kullanım yerlerini eğitimin ilerleyen bölümlerinde detaylı göreceğiz.

Tabii ki, lambda fonksiyonları da tek bir değer döndermek zorunda değildir:

````python
def sum_multiply(a, b):
    return a + b, a * b

sum_multiply(a = 3, b = 2)
````

````python
sum_multiply = lambda a, b: (a + b, a * b)
sum_multiply(a = 3, b = 2)
````

### Birinci-sınıf fonksiyon (first-class function)

Birinci-sınıf bir fonksiyon (first-class function), belirli bir fonksiyon türü değildir. Python'daki tüm fonksiyon birinci sınıf fonksiyonlardır. *_Belirli bir programlama dilinde fonksiyonların birinci sınıf olduğunu söylemek, diğer nesne türleri (tamsayılar veya dizgiler gibi) gibi kullanılabileceği ve manipüle edilebileceği anlamına gelir._* Yani, Python'da bir fonksiyon:

* Fonksiyonlar da birer nesnedir ve değişkenlere atanabilir.
* Bir döndürülen değer (return value) olarak kullanılabilir
* Parametre olarak kullanılabilir.
* Hash tabloları, listeler gibi veri yapılarında saklanabilir...
