# 2.Ders Notu

Python'da değişkenlere atama her değişken memory'de ayrı bir alan tutmaktansa sadece ortak alanı işaret eder mesela bir örnek:

```python
temp = 35
print("temp : ", temp)
sicaklik = 35
print("sicaklik:", sicaklik)

print("temp id = ", id(temp))
print("sicaklik id =",id(sicaklik))
```

Örnekte de görüldüğü gibi farklı değişkenler olmalarına rağmen aynı id'yi veriyor, bu da aynı değeri işaret etmeleri ile sağlanıyor ve böylece daha iyi bir bir memory yönetimini destekliyor.
