Her veri tipi her programlama dilinde olmayabilir.
Pythonda veri yapıları 2 ana dala ayrılır;
1. İlkel (Primitive)
    - Tamsayı(integer)
    - Kayan nokta(float)
    - Dizgi(string)
    - Boole(Boolean)
2. İlkel Olmayan (Non-primitive)
    1. Yerleşik(Built-in)
        - Liste (list)
        - Demet (Tuple)
        - Sözlük (Dictionary)
        - Küme (set)
    2. Kullanıcı tanımlı (user-defined)
        - Yığın (stack)
        - Kuyruk (Queue)
        - Bağlı liste (Linked list)
        - Ağaç (tree)
        - Grafik (Graph)

Programlama paradigmaları; programlama dilinin çözüm getirdiği problemlere karşı kullanmış olduğu yöntemin stili de denilebilir.
Programlama Paradigmaları:

![Programlama Paradigmaları](images/programlama-paradigmaları.png)

Python çoklu-paradigma yapısına uymaktadır, bu paradigmalar; 
- emirli(imperative) 
- prosedürel(procedural)
- fonksiyonel(functional)
- nesne-tabanlı programlama(object-oriented programming) 

olmak üzere dört ana paradigmayı destekleyen ve genel amaçlı programlama için tasarlanmış üst düzey bir programlama dilidir.

Emirli(imperative) paradigmada bir işlemi nasıl yapacağını anlattığın (how it will happen), bildirimili(declarative) ise bir işlemin ne yapacağı (sonuca odaklı) (what will happen) anlattığın programlama şeklidir.
